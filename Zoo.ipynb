```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(zoo.drop('type',1))
```

```python
scaler = scaler.transform(zoo.drop('type',1))
```

```python
df = pd.DataFrame(scaler,columns = zoo.columns[:-1])
```

```python
df.head()
```

```python
#sns.pairplot(data = zoo)

#plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain,xTest, yTrain,yTest = train_test_split(df,zoo.type,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(2)
```

```python
knn = knn.fit(xTrain,yTrain)
```

```python
yPred = knn.predict(xTest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
confusion_matrix(yTest,yPred)
```

```python
print(classification_report(yTest,yPred))
```

```python

from imblearn.over_sampling import SMOTE 
sm = SMOTE(k_neighbors=2) 

xTrain,yTrain = sm.fit_sample(xTrain,yTrain)
```

```python
from sklearn.model_selection import cross_val_score
acc=list()
for x in range(1,30):
    knn = KNeighborsClassifier(x)
    knn = cross_val_score(knn,xTrain,yTrain,cv=5)
    acc.append(knn.mean())
```

```python
plt.figure(figsize=(19,9))

plt.plot(range(1,30),acc)

plt.show()
```

```python
# so can conclude k value =11

knn = KNeighborsClassifier(11)

knn = knn.fit(xTrain,yTrain)
```

```python
yPred = knn.predict(xTest)
```

```python
accuracy_score(yTest,yPred)
```

```python
print(classification_report(yTest,yPred))
```
